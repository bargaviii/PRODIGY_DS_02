```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\Bargavi\\Desktop\\train.csv")  # Adjust the path if needed

# Calculate survival counts by gender
survival_by_gender = df.groupby('Sex')['Survived'].mean()

plt.figure(figsize=(6, 6))
plt.pie(
    survival_by_gender,
    labels=survival_by_gender.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=['skyblue', 'lightcoral']
)

plt.title('Survival Rate by Gender')
plt.axis('equal')  # Make it a circle
plt.show()
```

#### (The above pie chart shows the survival rate by gender)

#### (The above bar chart shows the survival of people)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = df.dropna(subset=['Age'])

bins = [0, 10, 20, 30, 40, 50, 60, 80]
labels = ['0â€“10', '11â€“20', '21â€“30', '31â€“40', '41â€“50', '51â€“60', '61+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Calculate survival rate by age group
age_survival = df.groupby('AgeGroup')['Survived'].mean().reset_index()

# Plot bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x='AgeGroup', y='Survived', data=age_survival, palette='inferno')

plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

#### (The above bar chart shows survival rate of each age groups)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Drop missing fare values
df = df.dropna(subset=['Fare'])

# Create fare bins (quartiles)
df['FareRange'] = pd.qcut(df['Fare'], 4, labels=['Low', 'Medium', 'High', 'Very High'])

# Group by FareRange and calculate survival rate
fare_survival = df.groupby('FareRange')['Survived'].mean().reset_index()

# Plot survival rate by fare range
plt.figure(figsize=(8, 5))
sns.barplot(x='FareRange', y='Survived', data=fare_survival, palette='flare')

# Customize the plot
plt.title('Survival Rate by Fare Range')
plt.xlabel('Fare Category')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

#### (The abov bar plot shows the survival rate by fare change)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style="whitegrid")

# Group data by gender and class, and calculate survival rate
grouped = df.groupby(['Sex', 'Pclass'])['Survived'].mean().reset_index()

# Plot grouped bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=grouped, palette='icefire')

# Customize plot
plt.title('Survival Rate by Gender and Passenger Class')
plt.xlabel('Passenger Class (1 = 1st, 2 = 2nd, 3 = 3rd)')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

#### (The above bar chart shows the survival rate by gender and passenger class)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Drop rows with missing 'Embarked' values
df = df.dropna(subset=['Embarked'])

# Group by embarkation point and calculate survival rate
embark_survival = df.groupby('Embarked')['Survived'].mean().reset_index()

# Plot bar chart
plt.figure(figsize=(6, 4))
sns.barplot(x='Embarked', y='Survived', data=embark_survival, palette='rocket')

# Customize chart
plt.title('Survival Rate by Embarkation Point')
plt.xlabel('Embarkation Point (C = Cherbourg, Q = Queenstown, S = Southampton)')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

#### (The above bar chart shows survival rate by embarkation point)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate family size
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Group by family size and calculate survival rate
family_survival = df.groupby('FamilySize')['Survived'].mean().reset_index()

# Plot survival rate vs family size
plt.figure(figsize=(8, 5))
sns.barplot(x='FamilySize', y='Survived', data=family_survival, palette='viridis')

# Customize the plot
plt.title('Survival Rate by Family Size')
plt.xlabel('Family Size (Siblings/Spouses + Parents/Children + Self)')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

#### (The above bar chart shows survival rate by family size)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("C:\\Users\\Bargavi\\Desktop\\train.csv")

# Create a 'FamilySize' column
df['FamilySize'] = df['SibSp'] + df['Parch']

# Create 'Alone' column: 1 if alone, 0 if with family
df['Alone'] = df['FamilySize'].apply(lambda x: 'Alone' if x == 0 else 'With Family')

# Set seaborn style
sns.set(style="whitegrid")

# Plot survival counts based on Alone status
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='Alone', hue='Survived', palette='pastel')

# Customize the plot
plt.title('Survival Count: Alone vs With Family')
plt.xlabel('Passenger Status')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

#### (The above bar chart shows the survivors and non survivors side by side for passengers who were alone and with families)
